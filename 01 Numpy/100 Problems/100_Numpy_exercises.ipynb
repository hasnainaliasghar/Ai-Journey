#### 63. Create an array class that has a name attribute (★★☆)

````python
class NamedArray(np.ndarray):
    def __new__(cls, array, name="no name"):
        obj = np.asarray(array).view(cls)
        obj.name = name
        return obj

    def __array_finalize__(self, obj):
        if obj is None:
            return
        self.name = getattr(obj, 'name', 'no name')

Z = NamedArray(np.arange(10), "range_10")
print(Z.name, Z)
````

#### 75. How to compute averages using a sliding window over an array? (★★★)

````python
def sliding_window_mean(arr, window):
    cumsum = np.cumsum(arr)
    cumsum[window:] = cumsum[window:] - cumsum[:-window]
    return cumsum[window - 1:] / window

Z = np.arange(10, dtype=float)
print(sliding_window_mean(Z, 3))
````

#### 79. Consider 2 sets of points P0,P1 describing lines (2d) and a set of points P, how to compute distance from each point j (P[j]) to each line i (P0[i],P1[i])? (★★★)

````python
P0 = np.random.uniform(-1, 1, (3, 2))
P1 = np.random.uniform(-1, 1, (3, 2))
P  = np.random.uniform(-1, 1, (4, 2))

# D[j, i] = distance from P[j] to line (P0[i], P1[i])
line_vecs = P1 - P0  # (3, 2)
norms = np.linalg.norm(line_vecs, axis=1)  # (3,)
# cross product (scalar in 2D) for each combination
D = np.abs(np.cross(line_vecs[np.newaxis, :, :],
                    P0[np.newaxis, :, :] - P[:, np.newaxis, :])) / norms
print(D.shape)  # (4, 3)
````

#### 80. Consider an arbitrary array, write a function that extracts a subpart with a fixed shape and centered on a given element (pad with a `fill` value when necessary) (★★★)

````python
def extract_subpart(arr, center, shape, fill=0):
    """Extract a subpart of arr of given shape centered on center, padding with fill."""
    center = np.array(center)
    shape  = np.array(shape)
    pad_width = [(s // 2, s // 2) for s in shape]
    padded = np.pad(arr, pad_width, mode='constant', constant_values=fill)
    slices = tuple(slice(c, c + s) for c, s in zip(center, shape))
    return padded[slices]

Z = np.arange(25).reshape(5, 5)
print(extract_subpart(Z, (2, 2), (3, 3)))
````

#### 85. Create a 2D array subclass such that Z[i,j] == Z[j,i] (★★★)

````python
class SymmetricArray(np.ndarray):
    def __setitem__(self, index, value):
        i, j = index
        super().__setitem__((i, j), value)
        super().__setitem__((j, i), value)

n = 5
Z = np.zeros((n, n)).view(SymmetricArray)
Z[1, 3] = 42
print(Z[3, 1])  # should be 42
````

#### 88. How to implement the Game of Life using numpy arrays? (★★★)

````python
def game_of_life(Z):
    """One step of Conway's Game of Life."""
    # Count live neighbours using convolution-like sum
    N = (Z[0:-2, 0:-2] + Z[0:-2, 1:-1] + Z[0:-2, 2:] +
         Z[1:-1, 0:-2]                  + Z[1:-1, 2:] +
         Z[2:,   0:-2] + Z[2:,   1:-1] + Z[2:,   2:])
    # Birth: dead cell with exactly 3 neighbours
    # Survival: live cell with 2 or 3 neighbours
    birth    = (N == 3) & (Z[1:-1, 1:-1] == 0)
    survival = ((N == 2) | (N == 3)) & (Z[1:-1, 1:-1] == 1)
    Z[1:-1, 1:-1] = birth | survival
    return Z

Z = np.random.randint(0, 2, (20, 20))
for _ in range(5):
    Z = game_of_life(Z)
print(Z)
````

#### 90. Given an arbitrary number of vectors, build the cartesian product (every combination of every item) (★★★)

````python
def cartesian_product(*arrays):
    grids = np.meshgrid(*arrays, indexing='ij')
    return np.stack(grids, axis=-1).reshape(-1, len(arrays))

result = cartesian_product(np.array([1, 2, 3]),
                           np.array([4, 5]),
                           np.array([6, 7]))
print(result)
````

#### 93. Consider two arrays A and B of shape (8,3) and (2,2). How to find rows of A that contain elements of each row of B regardless of the order of the elements in B? (★★★)

````python
A = np.random.randint(0, 5, (8, 3))
B = np.random.randint(0, 5, (2, 2))

# For each row in B, check if all elements exist in each row of A
rows = []
for b_row in B:
    mask = np.all(np.isin(b_row, A), axis=1) if A.ndim == 2 else np.isin(b_row, A).all()
    rows.append(np.where(np.array([np.all(np.isin(b_row, a_row)) for a_row in A]))[0])

print("A:\n", A)
print("B:\n", B)
print("Matching rows for each B row:", rows)
````

#### 97. Considering 2 vectors A & B, write the einsum equivalent of inner, outer, sum, and mul function (★★★)

````python
A = np.array([1, 2, 3])
B = np.array([4, 5, 6])

print("Inner (dot):", np.einsum('i,i->', A, B))       # == np.inner(A, B)
print("Outer:\n",     np.einsum('i,j->ij', A, B))     # == np.outer(A, B)
print("Sum:",         np.einsum('i->', A))             # == np.sum(A)
print("Mul:",         np.einsum('i,i->i', A, B))      # == A * B
````

#### 98. Considering a path described by two vectors (X,Y), how to sample it using equidistant samples (★★★)?

````python
# Build a path and sample at equal arc-length intervals
phi = np.arange(0, 10 * np.pi, 0.1)
X   = np.cos(phi)
Y   = np.sin(phi)

# Compute cumulative arc length
dX = np.diff(X)
dY = np.diff(Y)
lengths = np.cumsum(np.sqrt(dX**2 + dY**2))
lengths = np.insert(lengths, 0, 0)

# Sample at evenly spaced distances
n_samples = 50
sample_lengths = np.linspace(0, lengths[-1], n_samples)
Xs = np.interp(sample_lengths, lengths, X)
Ys = np.interp(sample_lengths, lengths, Y)
print("Sampled", n_samples, "equidistant points")
````

#### 100. Compute bootstrapped 95% confidence intervals for the mean of a 1D array X (i.e., resample the elements of an array with replacement N times, compute the mean of each sample, and then compute percentiles over the means). (★★★)

````python
X = np.random.randn(100)
N = 1000  # number of bootstrap samples

# Resample with replacement N times and compute means
bootstrap_means = np.mean(
    X[np.random.randint(0, len(X), (N, len(X)))],
    axis=1
)

ci_low  = np.percentile(bootstrap_means, 2.5)
ci_high = np.percentile(bootstrap_means, 97.5)
print(f"95% CI: [{ci_low:.4f}, {ci_high:.4f}]")
````
