```python
import seaborn as sns
import pandas as pd
import  cufflinks as cf
from plotly.offline import iplot
import plotly.express as px
from seaborn import boxplot
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
fig = px.line(tips, y = 'total_bill')
fig.show()
```

```python
avg_tip = tips.groupby("day", as_index=False)["tip"].mean()
px.bar(avg_tip, x="day", y="tip")
```

```python
px.scatter(x = tips['total_bill'], y=tips['tip'], color=tips['day'])
```

```python
px.box(x = tips['total_bill'], y = tips['day'])
```
